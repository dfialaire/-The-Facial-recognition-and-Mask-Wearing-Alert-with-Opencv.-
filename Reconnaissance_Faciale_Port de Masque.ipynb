<H1> 2ème partie : Captures de photos d'identification d'un individu avec id défini </H1>
--> Lancez la cellule,
--> donnez un N° d'Id à la personne filmée pendant 20 secondes, Attention : le Nom doit être mis entre Guillemets !
--> La capture vidéo s'arrête toute seule. <br />
         ** Vérifiez dans votre dossier 'dataset' la qualité des captures de visages enregistrées (id correcte, un seul visage par id...) ** <br />

```python
# -*- coding: utf-8 -*-

import cv2
import sqlite3
import numpy as np

faceDetect = cv2.CascadeClassifier("haarcascade_frontalface_default.xml");
cam=cv2.VideoCapture(0);

def insertOrUpdate(Id, Name):
    conn=sqlite3.connect("FaceBase.db")
    cmd="SELECT * FROM People WHERE ID="+str(Id)
    cursor=conn.execute(cmd)
    isRecordExist=0
    for row in cursor:
        isRecordExist=1
    if(isRecordExist==1):
        cmd="UPDATE People SET Name "+srt(Name)+" WHERE ID="+srt(Id)
    else:
        cmd="INSERT INTO People(Id, Name) Values("+str(Id)+","+str(Name)+")"
    conn.execute(cmd)
    conn.commit()
    conn.close()

id=input("Entrez un N° d'id : ")
name=input("Entrez un nom : ")
insertOrUpdate(id, name)


sampleNum=0
while(True):
    ret,img=cam.read();
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=faceDetect.detectMultiScale(gray,1.3,5);
    for (x,y,w,h) in faces:
        sampleNum=sampleNum+1
        cv2.imwrite("dataSet/User."+str(id) +"."+str(sampleNum)+".jpg", gray[y:y+h, x:x+w])
        cv2.rectangle(img, (x,y), (x+w, y+h), (0, 255, 0), 2 )
        cv2.waitKey(100);
    cv2.imshow("Face", img);
    cv2.waitKey(1);
    if(sampleNum>100):
        break
cam.release()
cv2.destroyAllWindows()
```

<H1> 3ème partie : Entrainement de l'algorithme pour la détection des visages photographiés présents dans le dossier 'dataset' : </H1>
--> Lancez cette cellule obligatoirement avant de lancer la 3ème partie, et renouvellez si un nouveau visage est inséré dans la base de données. <br />
==> Vous devez normalement observer le défilement des photos capturées à votre écran ; <br />
==> A la fin, vérifiez que le fichier 'trainningData.yml' a bien été créé dans votre dossier 'recognizer'.


```python
# -*- coding: utf-8 -*-

import os
import cv2
import numpy as np
from PIL import Image
import sqlite3

#0# recognizer=cv2.createLBPHFaceRecognizer();
recognizer=cv2.face.LBPHFaceRecognizer_create();
#1# recognizer=cv2.face.LBPHFaceRecognizer_create();

path="dataSet"

def getImagesWithID(path):
    imagePaths=[os.path.join(path,f) for f in os.listdir(path)]
    faces=[]
    IDs=[]
    for imagePath in imagePaths:
        faceImg=Image.open(imagePath).convert("L");
        faceNp=np.array(faceImg, "uint8")
        ID=int(os.path.split(imagePath) [-1].split(".") [1])
        faces.append(faceNp)
        IDs.append(ID)
        cv2.imshow("training",faceNp)
        cv2.waitKey(10)
        
        #return np.array(IDs), faces
    return np.array(IDs), faces
    
Ids , faces = getImagesWithID(path)
recognizer.train(faces, Ids)
recognizer.save("recognizer/trainningData.yml")
cv2.destroyAllWindows()

```

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Feb 14 11:12:18 2020

@author: dfial
"""

import cv2
import numpy as np
import sqlite3

faceDetect = cv2.CascadeClassifier("haarcascade_frontalface_default.xml");
cam=cv2.VideoCapture(0);
rec=cv2.face.LBPHFaceRecognizer_create();
rec.read("recognizer/trainningData.yml")
path="dataSet"

def getProfile(id):
    conn=sqlite3.connect("FaceBase.db")
    cmd="SELECT * FROM People WHERE ID="+str(id)
    cursor=conn.execute(cmd)
    profile=None
    for row in cursor:
        profile=row
    conn.close()
    return profile


#id=[]
id=0

#font=cv2.cv.InitFont(cv2.cv.CV_FONT_HERSHEY_COMPLEX_SMALL, 5,1,0,4)
#font=cv2.FONT_HERSHEY_COMPLEX_SIMPLEX
font = cv2.FONT_HERSHEY_SIMPLEX


##    font = cv2.FONT_HERSHEY_SIMPLEX
##    cv2.putText(img,"Cat",(x,y-10),font,0.55,(0,255,0),1)

while(True):
    ret,img=cam.read();
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces=faceDetect.detectMultiScale(gray,1.3,5);
    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y), (x+w, y+h), (0, 255, 0), 2 )
        id,conf=rec.predict(gray[y:y+h, x:x+w])
       
        if conf>70:          # Plus la valeur de conf est élevé, moins l'algorithme a reconnu des visages
            id = "inconnu {0:.0f}%".format(round(100 - conf, 2)) # Cela évite qu'il affiche un prénom avec un % négatif
            cv2.putText(img,str(id),(x,y+h), font, 1, (255,255,255), 3);
        else:
            profile=getProfile(id)
            if (profile!=None):

            #cv2.cv.putText(cv2.cv.fromarray(img),str(id),(x,y+h), font, 255);
                cv2.putText(img,str(profile[1]),(x,y+h+30), font, 1, (255,255,255), 3);
                cv2.putText(img,str(profile[2]),(x,y+h+60), font, 1, (255,255,255), 3);
                cv2.putText(img,str(profile[3]),(x,y+h+90), font, 1, (255,255,255), 3);
                cv2.putText(img,str(profile[4]),(x,y+h+120), font, 1, (255,255,255), 3);

                ##  cv2.putText(im, str(Id), (x,y-40), font, 1, (255,255,255), 3)

    cv2.imshow("En direct _ Video Webcam", img);
    
    if(cv2.waitKey(1) ==ord("q")):
        break;
cam.release()
cv2.destroyAllWindows()
```
